Abstract:
This data represents sales transactions across different regions managed by various sales managers and salesmen. It includes information such as the region, manager, salesman, item sold, units sold, unit price, and total sale amount. The analysis aims to identify patterns, trends, and insights from the data to understand the performance and dynamics of sales within each region.

```python
import pandas as pd

github_url = 'https://raw.githubusercontent.com/JuberMohammad2411/salesdata/main/salesdata.csv'

df = pd.read_csv(github_url)

print(df)




region_sales = df.groupby('Region')['Sale_amt'].sum()
region_sales.plot(kind='bar', xlabel='Region', ylabel='Total Sales Amount', title='Total Sales Amount by Region')
plt.show()


manager_units = df.groupby('Manager')['Units'].sum()
manager_units.plot(kind='bar', xlabel='Manager', ylabel='Units Sold', title='Units Sold by Manager')
plt.show()


item_sales = df.groupby('Item')['Sale_amt'].sum()
item_sales.plot(kind='pie', ylabel='Sales Amount', title='Sales Amount Distribution by Item')
plt.show()


salesman_avg_price = df.groupby('SalesMan')['Unit_price'].mean()
salesman_avg_price.plot(kind='bar', xlabel='Salesman', ylabel='Average Unit Price', title='Average Unit Price by Salesman')
plt.show()

top_items = df.groupby('Item')['Sale_amt'].sum().nlargest(5)
top_items.plot(kind='bar', xlabel='Item', ylabel='Total Sales Amount', title='Top Selling Items by Sales Amount')
plt.show()

region_item_units = df.groupby(['Region', 'Item'])['Units'].sum().unstack()
region_item_units.plot(kind='bar', stacked=True, xlabel='Region', ylabel='Units Sold', title='Units Sold by Region and Item Category')
plt.show()


item_avg_price = df.groupby('Item')['Unit_price'].mean()
item_avg_price.plot(kind='barh', xlabel='Average Unit Price', ylabel='Item', title='Average Unit Price by Item Category')
plt.show()


manager_sales_contribution = df.groupby('Manager')['Sale_amt'].sum()
manager_sales_contribution.plot(kind='pie', ylabel='Sales Amount', title='Sales Amount Contribution by Manager')
plt.show()


top_salesmen = df.groupby('SalesMan')['Sale_amt'].sum().nlargest(5)
top_salesmen.plot(kind='bar', xlabel='Salesman', ylabel='Total Sales Amount', title='Top Salesmen by Sales Amount')
plt.show()


item_sales_distribution = df.groupby('Item')['Units'].sum()
item_sales_distribution.plot(kind='pie', ylabel='Units Sold', title='Sales Distribution by Item Category')
plt.show()
```
