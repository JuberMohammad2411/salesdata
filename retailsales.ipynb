```python

Abstract:
This dataset comprises information on sales transactions categorized by product type. The dataset includes columns such as Net Quantity
(the quantity of each product sold), Gross Sales (the total sales revenue before any deductions), Discounts (the amount deducted from the gross 
sales due to discounts), Returns (the quantity of products returned), and Total Net Sales (the final revenue after accounting for discounts and returns)
.This abstract aims to analyze the dataset and draw insights into the sales performance of different product types.
```

```python
import mysql.connector
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Connect to MySQL database
conn = mysql.connector.connect(host="localhost", user="root", passwd="root", database="retailsales")

# Define SQL query
query = "SELECT * FROM busines_retailsales"

# Read data into Pandas DataFrame
df = pd.read_sql(query, conn)

# Display the DataFrame
print(df.head())

# Close the connection
conn.close()
```

```python
# Which product type has the highest total net sales?
# Calculate total net sales for each product type
total_net_sales_per_type = df.groupby('Product Type')['Total Net Sales'].sum()

# Plotting
plt.figure(figsize=(10, 6))
total_net_sales_per_type.plot(kind='bar')
plt.title('Total Net Sales by Product Type')
plt.xlabel('Product Type')
plt.ylabel('Total Net Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate average gross sales for each product type
avg_gross_sales_per_type = df.groupby('Product Type')['Gross Sales'].mean()

# Plotting
plt.figure(figsize=(10, 6))
avg_gross_sales_per_type.plot(kind='bar')
plt.title('Average Gross Sales by Product Type')
plt.xlabel('Product Type')
plt.ylabel('Average Gross Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plotting
plt.figure(figsize=(10, 6))
sns.boxplot(x='Product Type', y='Discounts', data=df)
plt.title('Distribution of Discounts by Product Type')
plt.xlabel('Product Type')
plt.ylabel('Discounts')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate total returns for each product type
total_returns_per_type = df.groupby('Product Type')['Returns'].sum()

# Plotting
plt.figure(figsize=(10, 6))
total_returns_per_type.plot(kind='bar')
plt.title('Total Returns by Product Type')
plt.xlabel('Product Type')
plt.ylabel('Total Returns')
plt.xticks(rotation=45)
plt.show()
```

```python
#What is the correlation between gross sales and discounts?
# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(df['Gross Sales'], df['Discounts'])
plt.title('Correlation between Gross Sales and Discounts')
plt.xlabel('Gross Sales')
plt.ylabel('Discounts')
plt.show()
```

```python
# Calculate proportion of net quantity sold for each product type
total_quantity_sold = df['Net Quantity'].sum()
proportion_sold_per_type = (df.groupby('Product Type')['Net Quantity'].sum() / total_quantity_sold) * 100

# Plotting
plt.figure(figsize=(10, 6))
proportion_sold_per_type.plot(kind='pie', autopct='%1.1f%%')
plt.title('Proportion of Net Quantity Sold by Product Type')
plt.ylabel('')
plt.show()
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.hist(df['Total Net Sales'], bins=20, edgecolor='black')
plt.title('Distribution of Total Net Sales')
plt.xlabel('Total Net Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
# Calculate average net quantity sold for each product type
avg_net_quantity_per_type = df.groupby('Product Type')['Net Quantity'].mean()

# Plotting
plt.figure(figsize=(10, 6))
avg_net_quantity_per_type.plot(kind='bar')
plt.title('Average Net Quantity Sold by Product Type')
plt.xlabel('Product Type')
plt.ylabel('Average Net Quantity Sold')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate average returns per unit sold for each product type
avg_returns_per_unit_sold = (df.groupby('Product Type')['Returns'].sum() / df.groupby('Product Type')['Net Quantity'].sum())

# Plotting
plt.figure(figsize=(10, 6))
avg_returns_per_unit_sold.plot(kind='bar')
plt.title('Average Returns per Unit Sold by Product Type')
plt.xlabel('Product Type')
plt.ylabel('Average Returns per Unit Sold')
plt.xticks(rotation=45)
plt.show()
```

```python
#What is the distribution of total net sales for each product type?
# Plotting
plt.figure(figsize=(10, 6))
sns.boxplot(x='Product Type', y='Total Net Sales', data=df)
plt.title('Distribution of Total Net Sales by Product Type')
plt.xlabel('Product Type')
plt.ylabel('Total Net Sales')
plt.xticks(rotation=45)
plt.show()
```
